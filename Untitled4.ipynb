```python
## Region correction according to states
nf.loc[(nf.State == "Delhi") | (nf.State == "Uttar Pradesh") |
        (nf.State == "Haryana") | (nf.State == "HP") | (nf.State == "J&K")| (nf.State == "MP"), "Region"] = "North"

nf.loc[(nf.State == "Andhra Pradesh") | (nf.State == "Karnataka") |
        (nf.State == "Kerala")  | (nf.State == "Tamilnadu") | 
        (nf.State == "Telengana"), "Region"] = "South"

nf.loc[(nf.State == "Assam") | (nf.State == "Jharkhand") |
        (nf.State == "Tripura") | (nf.State == "West Bengal"), "Region"] = "East"

nf.loc[(nf.State == "Gujarat"), "Region"] = "West"
        
nf.loc[(nf.State == "Bihar") , "Region"] = "North East"

nf.loc[(nf.State == "Goa") | (nf.State == "Maharashtra"), "Region"] = "South West"

nf.loc[(nf.State == "Odisha"), "Region"] = "South East"
        
nf.loc[(nf.State == "Rajasthan"), "Region"] = "North West"     

```
